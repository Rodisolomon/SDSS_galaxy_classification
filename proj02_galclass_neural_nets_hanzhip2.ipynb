```python
import numpy as np
from PIL import Image
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
def rgb_to_grayscale(img):
    """
    convert a rgb image to a grayscale image, return a np array
    (width, height, 3)->(width, height)
    """
    gray_img = 0.21 *img[:,:,0] + 0.72 * img[:,:,1] + 0.07 * img[:,:,2]
    return gray_img

def open_img(gal_id, path):
    """
    gal_id - a string of galaxy
    path - string
    return a (424, 424, 3) np array
    """
    img_file = path+str(gal_id)+'.jpg'
    im = Image.open(img_file)
    pixels = np.asarray(im)
    im.close()
    return pixels
# def whole_transform(df, path, width = 424, height = 424):
#     """
#     accept df, use the array of galaxy id, convert from gdb to grayscale image
#     return an (num, width-424, height-424) numpy array & num of galaxy in this list
#     """
#     num = np.shape(df)[0]
#     gray_galaxies = np.empty([num, width, height])
#     for index, galaxy in enumerate(df):
#         gal_id = df['GalaxyID'][index]
#         rgb_img = open_img(gal_id, path)
#         gray_img = rgb_to_grayscale(rgb_img)
#         gray_galaxies[index] = gray_img

#     return gray_galaxies, num
```

```python
# from PIL import Image
#Similar to the sample given by MNIST dataset, I'm going to use 20% of data as test sample, 80% as training.
#In this 80% I'll use 10% for validation set and 70% for training data
import random
def whole_transform(df, path):
    """
    accept df, use the array of galaxy id, convert from gdb to grayscale image
    return an (num, width-424, height-424) numpy array & num of galaxy in this list
    """
    num = np.shape(df)[0]
    train_path_el = '/Users/helen/AA211/final_project_data/train/el/'
    train_path_sp = '/Users/helen/AA211/final_project_data/train/sp/'
    test_path_el = '/Users/helen/AA211/final_project_data/test/el/'
    test_path_sp = '/Users/helen/AA211/final_project_data/test/sp/'
    vali_path_el = '/Users/helen/AA211/final_project_data/validation/el/'
    vali_path_sp = '/Users/helen/AA211/final_project_data/validation/sp/'
    train_id, test_id, vali_id = [], [], []
    for index in range(num - 1):
        flag = random.randint(0, 9)
        gal_id = df['GalaxyID'][index]
        rgb_img = open_img(gal_id, path)
        gray_img = rgb_to_grayscale(rgb_img)
        gray_img = gray_img.astype(np.uint8)
        gray_img = gray_img[90:335, 90:335] #(245, 245)
        result = Image.fromarray((gray_img * 255))
        if (result.mode != 'RGB'):
            result = result.convert('RGB')
        #split into three set
        if flag < 2: #test
            test_id.append(index)
            if df['Class1.1'][index] > 0.5:
                result.save(test_path_el + str(gal_id)+'.jpg')
            else:
                result.save(test_path_sp + str(gal_id)+'.jpg')
        elif (flag >= 2) and (flag <= 8): #train
            train_id.append(index)
            if df['Class1.1'][index] > 0.5:
                result.save(train_path_el + str(gal_id)+'.jpg')
            else:
                result.save(train_path_sp + str(gal_id)+'.jpg')
        else: #validation
            vali_id.append(index)
            if df['Class1.1'][index] > 0.5:
                result.save(vali_path_el + str(gal_id)+'.jpg')
            else:
                result.save(vali_path_sp + str(gal_id)+'.jpg')

    train_id, test_id, vali_id = np.array(train_id), np.array(test_id), np.array(vali_id)
    return train_id, test_id, vali_id
```
